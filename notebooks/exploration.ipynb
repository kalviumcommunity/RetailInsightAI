# Retail Customer Segmentation — Exploratory Analysis

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.preprocessing import load_and_clean, build_rfm
from src.clustering import train_clustering
from src.analytics import calculate_clv, cluster_summary
from src.recommendation import get_top_products_per_cluster
```

## 1. Load and Clean Data

```python
df = load_and_clean('../data/Online Retail.csv')
print(f'Rows: {len(df)}, Customers: {df["CustomerID"].nunique()}')
df.head()
```

## 2. Dataset Info and Missing Values

```python
df.info()
print('\nMissing values:')
print(df.isnull().sum())
```

## 3. Build RFM Features

```python
rfm = build_rfm(df)
rfm.describe()
```

## 4. RFM Distributions

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for ax, col in zip(axes, ['Recency', 'Frequency', 'Monetary']):
    sns.histplot(rfm[col], bins=40, ax=ax)
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

## 5. Train Clustering

```python
rfm = train_clustering(rfm)
rfm['Cluster'].value_counts().sort_index()
```

## 6. CLV and Cluster Summary

```python
rfm = calculate_clv(rfm)
cluster_summary(rfm)
```

## 7. Cluster Heatmap

```python
summary = cluster_summary(rfm)
plt.figure(figsize=(8, 3))
sns.heatmap(summary, annot=True, fmt='.1f', cmap='Blues', linewidths=0.5)
plt.title('Cluster Behaviour Heatmap')
plt.show()
```

## 8. Top Products per Cluster

```python
top_products = get_top_products_per_cluster(df, rfm)
top_products
```
